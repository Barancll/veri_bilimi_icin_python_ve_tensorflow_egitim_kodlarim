```python
yasListesi = [10,20,30,30,30,40,50,60,70,75]
```

```python
kiloListesi = [20,60,80,85,86,87,70,90,95,90]
```

```python
numpyYasListesi = np.array(yasListesi)
numpyKiloListesi = np.array(kiloListesi)
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"y")
plt.title("Kilonun Yaşa Göre Değişim Grafiği")
plt.xlabel("Yas")
plt.ylabel("Kilo")
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1** 3
```

```python
numpyDizisi2
```

Özelleştirme

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g")
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g--")
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g*-")
```

```python
plt.subplot(1,2,1)
plt.plot(numpyDizisi1,numpyDizisi2,"r*-")
plt.subplot(1,2,2) #1 sıra olcak, 2 kolon olcak, şuan 2. grafiği çiziyorum
plt.plot(numpyDizisi2,numpyDizisi1,"b--")
```

```python
benimFigure = plt.figure()
figureAxes = benimFigure.add_axes([0.4,0.4,0.6,0.6])
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("X ekseni veri ismi")
figureAxes.set_ylabel("Y ekseni veri ismi")
figureAxes.set_title("Grafik Başlığı")
```

SUBPLOT

```python
figure2 = plt.figure()
    
eksen1 = figure2.add_axes([0.1,0.1,0.9,0.9])
eksen2 = figure2.add_axes([0.3,0.4,0.3,0.3])

eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("X ekseni")
eksen1.set_ylabel("Y ekseni")
eksen1.set_title("Ana Grafik Başlığı")

eksen2.plot(numpyDizisi2,numpyDizisi1,"b")
eksen2.set_xlabel("X ekseni")
eksen2.set_ylabel("Y ekseni")
eksen2.set_title("JR. Grafik Başlığı")
```

```python
(benimFigure, benimEksen) = plt.subplots(nrows=1,ncols=2)
type(benimEksen)

for eksen in benimEksen:
    eksen.plot(numpyDizisi1,numpyDizisi2,"g")
    eksen.set_xlabel("X Ekseni")
```

```python
yeniFigur = plt.figure(figsize=(4,4))
#dpi=150
yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])
yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 2,label = "numpyDizisi ** 2")
yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 3,label = "numpyDizisi ** 3")
yeniEksen.legend(loc=6)
```

matplotlib kayıt yapmak
##dpi = kalite parametresi

```python
yeniFigur.savefig("benimfigur.png", dpi=200)
```

```python
(yeniFigur, yeniEksen) = plt.subplots()
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 2, color = "blue", linewidth = 4.0)
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 3, color = "yellow")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 4, color = "orange", linestyle = "--")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 5, color = "red", linestyle = ":")
```
